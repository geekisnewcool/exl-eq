```python
from math import sin, cos, sqrt, atan2, radians
Store_Name = []
Store_Name_1 = []
Shopper_id = []
minimum_distance = []
R = 6373.0
matrix=[]
for i in range(141,211):
    maxx = 500000
    name='Shopper_'+str(i)
#     print "name",name
    Shopper_id.append(name)
    print "Shopper_",i
    e= data_a['Max-Floor-visited'][i-1].strip(" ").split()
#     print "e",e
    e1= data_a['Max-Floor-visited'][i-1].strip(" ")
#    print "max_floor",e2
    if len(e) == 1:
#         print "in if len(e) == 1"
        shop_list=[]
        s = bigdata[bigdata['Shopper_ID']=='Shopper_'+str(i)].groupby(['Floors']).get_group(e[0])['lat']
#         print "s",s
        length = len(s)
        latt = df1[df1['Floor_Index']== int(e[0])]
        length_shops = latt['Store_Name'].count()
#         print "length of shops", length_shops
        for j in range(length_shops):
#             print "j",j
            sum_distance=0
            count_total=0
            lat1 = df1[df1['Floor_Index']== int(e[0])].reset_index()['latitude'][j]
            lon1 = df1[df1['Floor_Index']== int(e[0])].reset_index()['longitude'][j]
            for k in range(length):
#                 print "k" ,k 
                s = bigdata[bigdata['Shopper_ID']=='Shopper_'+str(i)].groupby(['Floors']).get_group(e[0])
                s_1 = pd.DataFrame(s).reset_index()
                lat2 = s_1['lat'][k]
                lon2 = s_1['lng'][k]
                
                dlon = lon2 - lon1
                dlat = lat2 - lat1

                a_1 = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
                c = 2 * atan2(sqrt(a_1), sqrt(1 - a_1))

                distance = R * c

                sum_distance = sum_distance + distance
                count_total = count_total + 1
            avg_sum_distance = sum_distance/count_total
#             print("Result:", avg_sum_distance)
            shop_list.append(avg_sum_distance)
            if avg_sum_distance < maxx :
#                 print "in if maxx"
                maxx = avg_sum_distance
                print "maxx",maxx
                Shop_name = df1[df1['Floor_Index']== int(e[0])].reset_index()['Store_Name'][j]
                Shop_name_1 = df1[df1['Floor_Index']== int(e[0])].reset_index()['Store_Name'][j] + str(j)
        Store_Name.append(Shop_name)
        Store_Name_1.append(Shop_name_1)
        minimum_distance.append(maxx)
        matrix.append(shop_list)
    if len(e) == 2:
#         print "in if e====2"
        shop_list=[]
        for d in range(len(e)):
            maxx1 = 500000
#             print "d",d
            s = bigdata[bigdata['Shopper_ID']=='Shopper_'+str(i)].groupby(['Floors']).get_group(e[d])['lat']
    #         print "s",s
            length = len(s)
            latt = df1[df1['Floor_Index']== int(e[d])]
            length_shops = latt['Store_Name'].count()
            print "length of shops", length_shops
            for j in range(length_shops):
#                 print "j",j
                sum_distance=0
                count_total=0
                lat1 = df1[df1['Floor_Index']== int(e[d])].reset_index()['latitude'][j]
                lon1 = df1[df1['Floor_Index']== int(e[d])].reset_index()['longitude'][j]
                for k in range(length):
#                     print "k" ,k 
                    s = bigdata[bigdata['Shopper_ID']=='Shopper_'+str(i)].groupby(['Floors']).get_group(e[d])
                    s_1 = pd.DataFrame(s).reset_index()
                    lat2 = s_1['lat'][k]
                    lon2 = s_1['lng'][k]

                    dlon = lon2 - lon1
                    dlat = lat2 - lat1
                    a_1 = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
                    c = 2 * atan2(sqrt(a_1), sqrt(1 - a_1))
                    distance = R * c
                    sum_distance = sum_distance + distance
                    count_total = count_total + 1
                avg_sum_distance = sum_distance/count_total
#                 print("Result:", avg_sum_distance)
                shop_list.append(avg_sum_distance)
                if avg_sum_distance < maxx1 :
#                     print "in if maxx"
                    maxx1 = avg_sum_distance
                    print "maxx",maxx1
                    Shop_name = df1[df1['Floor_Index']== int(e[d])].reset_index()['Store_Name'][j]
                    Shop_name_1 = df1[df1['Floor_Index']== int(e[d])].reset_index()['Store_Name'][j] + str(j)
        Store_Name.append(Shop_name)
        Store_Name_1.append(Shop_name_1)
        minimum_distance.append(maxx1)
        matrix.append(shop_list)
    
        
        
        
```
